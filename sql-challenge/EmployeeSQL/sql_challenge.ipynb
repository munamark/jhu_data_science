```python
from sqlalchemy import create_engine
from config import password
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/pewlett_hackard_employees')
```

```python
connection = engine.connect()
```

```python
salary_df = pd.read_sql("select * from salaries",con=engine)
#salary_df
```

```python
plt.figure(figsize=(11,7))
sns.histplot(data=salary_df, x="salary", binwidth = 10000)
plt.xlabel("Salary Range", size=13)
plt.ylabel("Number of Employees", size=13)
plt.title("Histogram of Employee Salaries", size=17)
```

```python
salary2_df = pd.read_sql("select title.title, round(avg(salaries.salary),2) as avg_salary from salaries join employees on salaries.emp_no = employees.emp_no join title on title.title_id = employees.emp_title_id group by 1",con=engine)
#salary2_df
```

```python
plt.figure(figsize=(11,7))
sns.barplot(x='avg_salary', y='title', data=salary2_df)
plt.xlabel("Average Salary", size=13)
plt.ylabel("Employee Title", size=13)
plt.title("Average Employee Salary by Title", size=17)
```
